```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import pandas as pd

# Setup Chrome WebDriver
driver_path = r"C:\chromedriver-win64\chromedriver-win64\chromedriver.exe"  # Your path to ChromeDriver
options = Options()
options.add_argument("--headless")  # Runs in headless mode (no GUI)
driver_service = Service(driver_path)

# Open Chrome browser with Selenium
driver = webdriver.Chrome(service=driver_service, options=options)

# URL for Yellow Pages search results for IT services in Durham, NC
url = "https://www.yellowpages.com/search?search_terms=IT+Service&geo_location_terms=Durham%2C+NC"
driver.get(url)

# Wait for the page to load
time.sleep(3)  # Adjust the wait time if necessary

# Extract page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Initialize a list to store company data
company_data = []

# Find all the companies in the search results
business_list = soup.find_all('div', class_='result')

for business in business_list:
    # Extract the company name
    company_name = business.find('a', class_='business-name').text.strip() if business.find('a', class_='business-name') else 'N/A'

    # Extract the industry/category
    industry_category = business.find('div', class_='categories').text.strip() if business.find('div', class_='categories') else 'N/A'

    # Extract the website URL
    website_section = business.find('a', class_='track-visit-website')
    website_url = website_section['href'] if website_section else 'N/A'

    # Extract the contact number
    contact_number = business.find('div', class_='phones').text.strip() if business.find('div', class_='phones') else 'N/A'

    # Extract the location/address
    location = business.find('div', class_='locality').text.strip() if business.find('div', class_='locality') else 'Durham, NC 27707'

    # Extract the company description
    description = business.find('div', class_='snippet').text.strip() if business.find('div', class_='snippet') else 'N/A'

    # Debug prints to inspect what's being extracted
    print(f"Company Name: {company_name}")
    print(f"Industry/Category: {industry_category}")
    print(f"Website URL: {website_url}")
    print(f"Contact Number: {contact_number}")
    print(f"Location: {location}")
    print(f"Description: {description}")
    print("-" * 40)

    # Append the data to the company_data list
    company_data.append([company_name, industry_category, website_url, contact_number, location, description])

# Close the WebDriver
driver.quit()

# Convert the data into a DataFrame
df = pd.DataFrame(company_data, columns=['Company Name', 'Industry/Category', 'Website URL', 'Contact Number', 'Location/Address', 'Company Description'])

# Save the data into a CSV file
df.to_csv('yellow_pages_it_services.csv', index=False)

print("Data has been scraped and saved to 'yellow_pages_it_services.csv'.")
```
